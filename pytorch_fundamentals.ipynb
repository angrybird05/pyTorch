```python
   ## finding min, max,mean,sum,etc(tensor aggregation)
import torch
x=torch.arange(0,100,10)
x
print(f"min: {x.min()}")
print(f"max: {x.max()}")
print(f"mean: {x.type(torch.float32).mean()}")
print(f"sum: {x.sum()}")
print(f"index of min value: {x.argmin()}")
print(f"index of max value: {x.argmax()}")
```

```python
    ## reshaping , stacking ,Squeezing and unsqueezing tensors
import torch

    x=torch.arange(1.,10.)
    print(f"orginal Tensor is {x},the shape of the orginal tensor is {x.shape}")
    x_reshaped = x.reshape(1,9)
    print(f"Reshaped Tensor{x_reshaped,x_reshaped.shape}")
    print("\n")


    #stacking a Tensor
    print("stacking the tensor\n")
    x_stacked = torch.stack([x,x,x,x],dim=0)
    print(f"stacked Tensor is {x_stacked}")
    print("\n")

    #squeezing the tensor
    print("squeezing the tensor\n")
    x_reshaped = x.reshape(9,1)
    print(f"the squeezed tensor is {x_reshaped}")

    print("\n")

    #unsqueezing the tensor
    print("unsqueezing the tensor\n")
    x_unsqueezed = x_reshaped.squeeze()
    print(f"orginal tensor{x_reshaped}")
    print(f"the unsqueezed tensor is {x_unsqueezed}")

    print("\n")

    #changing the view of the tensor
    print("changing the view of the tensor\n")
    z=x.view(1,9)
    print(f"\n the view tensor is {z}")

    print("\n")

    # Permuting the tensor
    print("Permute the tensor\n")
    # Fix: Create a 3D tensor to match the permute operation
    x_original = torch.rand(size=(224,224,3),dtype=torch.float32) # Example: 224x224 image with 3 channels
    x_permuted = x_original.permute(2,0,1)
    print(f"original shape: {x_original.shape}")
    print(f"permuted shape: {x_permuted.shape}")
```

```python
## Indexing on the tensor
import torch
x=torch.rand(4,4)
print(f"the orginal tensor is {x}")
print(f"the second row of the tensor is: {x[1]}")
print(f"the second column of the tensor is: {x[:,1 ]}")
print(f"the last column of the tensor is: {x[:,-1]}")
print(f"the last row of the tensor is: {x[-1]}")
```
